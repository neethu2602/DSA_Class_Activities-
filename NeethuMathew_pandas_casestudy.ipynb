```python
#Q5
df.rename(columns={"Title": "Movie_Title"})
```

```python
#Q6
unique_genrs=df["Genre"].unique()
unique_genrs
```

```python
#Q7
df["Num_Genres"] = df["Genre"].apply(lambda x: len(str(x).split("|")))
df.head()
```

```python
#Q8
df[["Revenue (Millions)", "Metascore"]].isnull().sum()
df["Revenue (Millions)"] = df["Revenue (Millions)"].fillna(df["Revenue (Millions)"].median())
df["Metascore"] = df["Metascore"].fillna(df["Metascore"].mean())
```

```python
df[["Revenue (Millions)", "Metascore"]].isnull().sum()
```

```python
#Q9
top5_rating = df.sort_values(by="Rating", ascending=False).head(5)
print(top5_rating[["Title", "Rating"]])
```

```python
#Q10
highest_revenue = df.loc[df["Revenue (Millions)"].idxmax(), ["Title", "Revenue (Millions)"]]
lowest_revenue  = df.loc[df["Revenue (Millions)"].idxmin(), ["Title", "Revenue (Millions)"]]

print("\nHighest Revenue Movie:")
print(highest_revenue)

print("\nLowest Revenue Movie:")
print(lowest_revenue)
```

```python
#Q11
avg_rating_director = (
    df.groupby("Director")["Rating"]
      .mean()
      .reset_index()
      .sort_values(by="Rating", ascending=False)
)

print(avg_rating_director)  # top 10 directors by avg rating
```

```python
#Q12
movies_per_year = df.groupby("Year")["Title"].count().sort_index()
print(movies_per_year)
```

```python
#Q13
high_perf_movies = df[(df["Rating"] > 8) & (df["Revenue (Millions)"] > 100)]
print(high_perf_movies[["Title", "Rating", "Revenue (Millions)"]])
```

```python
#Q14
director_stats = df.groupby("Director").agg({"Rating": "mean","Revenue (Millions)": "mean"}).sort_values(by="Rating", ascending=False)

print(director_stats)
```

```python
#Q15
year_stats = df.groupby("Year").agg({"Title":"count","Rating":"mean"}).sort_values(by="Year")
print(year_stats)
```

```python
#Q16
df["Profitability"] = df["Revenue (Millions)"] / df["Runtime (Minutes)"]
top10_profitable = df.nlargest(10,"Profitability")[['Title','Profitability']]
print(top10_profitable)


```

```python
#Q17

df["Rating_Category"] = np.where(df["Rating"] < 6, "Poor",
                        np.where(df["Rating"] < 7.5, "Average", "Excellent"))
df.head()
```

```python
#Q18
def movie_score(row):
    return row["Rating"] * 10 + row["Revenue (Millions)"]
df["High_Score"] = df.apply(movie_score, axis=1)
df.head()
```
