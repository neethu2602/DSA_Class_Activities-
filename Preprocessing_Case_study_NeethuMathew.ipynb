```python
data["Date House was Sold"] = pd.to_datetime(data["Date House was Sold"], format="%d %B %Y")
```

```python
missing=data.isnull().sum()
for col in missing[missing>0].index:

  median_value=data[col].median()
  data.loc[:, col] = data[col].fillna(median_value)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
minmax_cols = ["Flat Area (in Sqft)", "Lot Area (in Sqft)",
               "Area of the House from Basement (in Sqft)",
               "Basement Area (in Sqft)",
               "Living Area after Renovation (in Sqft)",
               "No of Floors"]

mscaler = MinMaxScaler()
data[minmax_cols] = mscaler.fit_transform(data[minmax_cols])

standard_cols = ["No of Bedrooms", "No of Bathrooms",
                 "Overall Grade", "Age of House (in Years)",
                 "Latitude", "Longitude"]

sscaler = StandardScaler()
data[standard_cols] = sscaler.fit_transform(data[standard_cols])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
houselabel = LabelEncoder()
data['Waterfront View'] = houselabel.fit_transform(data['Waterfront View'])
data['Condition of the House'] = houselabel.fit_transform(data['Condition of the House'])
```

```python
data=data.drop('ID',axis=1)
```

```python
num_cols=data.select_dtypes(include=('float64','int64'))
```

```python
numerical_cols = data.select_dtypes(include=['int64', 'float64','int32']).columns.tolist()
if 'Sale Price' in numerical_cols:
  numerical_cols.remove('Sale Price')

for col in numerical_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    from scipy.stats.mstats import winsorize
    data[col] = winsorize(data[col], limits=[0.05, 0.05])
    #data[col] = data[col].clip(lower=lower_bound, upper=upper_bound)
```
